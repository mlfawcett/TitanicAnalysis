```python
import pandas as pd

location = "datasets/titanicMac.xls"
titanic_df = pd.read_excel(location)

titanic_df.tail(10)
```

```python
# Just looking at the information in the spreadsheet

TitanicLength = len(titanic_df)
TitanicShape = titanic_df.shape
TitanicColNames = titanic_df.columns
TitanicDataCounts = titanic_df.count()

print("No. Rows ", TitanicLength)
print("Cols and Rows ", TitanicShape)
print("Cols Names ", TitanicColNames)
print("Data Counts\n", TitanicDataCounts)
```

```python
# looking at age counts

titanic_df['age'].value_counts()
```

```python
# Who was the oldest and youngest on board?

Oldest = titanic_df['age'].max()
Youngest = titanic_df['age'].min()
print("Oldest was ", Oldest)
print("Youngest was ", Youngest)
```

```python
# Who was the oldest and youngest to survive?

titanic_survived_df = titanic_df.loc[titanic_df['survived'] == 1]
OldestSurvived = titanic_survived_df['age'].max()
YoungestSurvived = titanic_survived_df['age'].min()
print("Oldest survivor was ", OldestSurvived)
print("Youngest survivor was ", YoungestSurvived)
```

```python
# What was all of the infomation for the oldest survivor

titanic_df.loc[titanic_df['age'] >= 80]
```

```python
# What was all of the information for the youngest survivor

titanic_df.loc[titanic_df['age'] < .3]
```

```python
#  Data frame for all those that perished

titanic_perished = titanic_df.loc[titanic_df['survived'] == 0]
```
